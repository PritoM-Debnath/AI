# N Queen Attack Detection

```python
def drawBoard(P):
  N = len(P)
  for i in range(1,N+1):
    print("|", end="")
    for j in range(1,N+1):
      if (i,j) in P:
        ind = str(1+ P.index((i,j)))
        print(" "+ind+" |", end= "")
      else:
        print(" * |", end= "")
    print()
```

```python
from math import ceil
import random
def generate_positions(N):
    CL = list(range(0,N*N));
    P = []
    for i in range(N):
        p = random.choice(CL)
        CL.remove(p)
        r = ceil((p+1)/N)
        c = 1+p%N
        P.append((r,c))
        #print('Q{}: {},{}'.format(i+1,r,c))
    return P

generate_positions(4)
```

```python
N = 5
positions = generate_positions(N)
drawBoard(positions)
```

```python
# row attack
def rowAttacK(P):
  #count = 0
  N=len(P)
  for i in range(N):
    for j in range(i+1,N):
      if P[i][0] == P[j][0]:
        #print('row attack:', end = " " )
        #print(P[i], end = ", ")
        #print(P[j]) 
        #count = count + 1 
        return True
  #return count
  return False

rowAttacK(positions)   
```

```python
#column attack
def colAttacK(P):
  #count = 0
  N=len(P)
  for i in range(N):
    for j in range(i+1,N):
      if P[i][1] == P[j][1]:
        #print('col attack:', end = " " )
        #print(P[i], end = ", ")
        #print(P[j]) 
        #count = count + 1 
        return True
  #return count
  return False

colAttacK(positions)
```

```python
# diagonal atack
def diaAttacK(P):
  #count = 0
  N=len(P)
  for i in range(N):
    for j in range(i+1,N):
      if abs(P[i][1]-P[j][1]) == abs(P[i][0]-P[j][0]):
        #print('diagonal attack:', end = " " )
        #print(P[i], end = ", ")
        #print(P[j]) 
        #count = count + 1 
        return True
  #return count
  return False

diaAttacK(positions)
```

```python
# implement a method to print a state is valid or not
# input row and col (two lists)
f=0
while(True):
  p=generate_positions(7)
  if rowAttacK(p)== False and colAttacK(p)==False and diaAttacK(p)==False:
    print(f)
    drawBoard(p)
    break
  f+=1



```
